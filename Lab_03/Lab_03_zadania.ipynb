```python
import numpy as np

a = np.array([[3, 6],
              [0, 9]])

b = np.array([[-3, -9],
              [1, 3]])


def odwrotna_2x2(macierz: np.array):
    if macierz.shape != (2,2):
        raise ValueError("Podana macierz nie jest macierzą 2 na 2")
    wyznacznik = macierz[0][0] * macierz[1][1] - macierz[0][1] * macierz[1][0]
    pomocna = np.zeros((macierz.shape[0], macierz.shape[1]))
    odwrotna = np.zeros((macierz.shape[0], macierz.shape[1]))
    if wyznacznik == 0:
        raise ValueError("Wyznacznik dla tej macierzy wynosi 0 a nie może!")
    pomocna[0][0] = macierz[1][1]
    pomocna[0][1] = -macierz[1][0]
    pomocna[1][0] = -macierz[0][1]
    pomocna[1][1] = macierz[0][0]

    liczba_wierszy = pomocna.shape[0]
    liczba_kolumn = pomocna.shape[1]

    transported_matrix = np.zeros((liczba_kolumn, liczba_wierszy))
    for i in range(liczba_wierszy):
        for j in range(liczba_kolumn):
            transported_matrix[j][i] = pomocna[i][j]

    for i in range(liczba_wierszy):
        for j in range(liczba_kolumn):
            odwrotna[i][j] = transported_matrix[i][j] / wyznacznik

    return odwrotna


print("----------------------------Macierz A ------------------------------------")
print("Macierz a odwrotna według mojej funkcji: \n", odwrotna_2x2(a))
print("Macierz a odwrotna według funkcji z numpy: \n", np.linalg.inv(a))
print("----------------------------Macierz B ------------------------------------")
# print("Macierz b odwrotna według mojej funkcji: \n", odwrotna_2x2(b)) # WYZNACZIK WYNOSI 0
# print("Macierz b odwrotna według funkcji z numpy: \n", np.linalg.inv(b)) # WYZNACZIK WYNOSI 0
```

```python
import numpy as np

a = np.array([[1, 0, 5],
              [2, 7, 6],
              [8, 3, 2]])

d = np.array([[3, -1, 1],
              [5, 1, 4],
              [-1, 3, 2]])


def odwrotna_3x3(a: np.array):
    if a.shape[0] != a.shape[1]:
            raise ValueError("Podana macierz nie jest kwadratowa")
    elif a.shape[0] != 3 or a.shape[1] != 3:
        raise ValueError("Wprowadzona macierz nie jest macierzą 3 x 3")
    else:
        det = a[0][0]*a[1][1]*a[2][2] + a[0][1]*a[1][2]*a[2][0] + a[0][2]*a[1][0]*a[2][1] - a[0][2]*a[1][1]*a[2][0] - a[0][0] * a[1][2]*a[2][1] - a[0][1]*a[1][0]*a[2][2]

        if(det == 0):
            raise ValueError("Wyznacznik podanej macierzy jest równy 0!")

    odwrotna = np.zeros((a.shape[0], a.shape[1]))

    odwrotna[0][0] = a[1][1]*a[2][2] - a[1][2]*a[2][1]
    odwrotna[0][1] = a[0][2]*a[2][1] - a[0][1]*a[2][2]
    odwrotna[0][2] = a[0][1]*a[1][2] - a[0][2]*a[1][1]

    odwrotna[1][0] = a[1][2]*a[2][0] - a[1][0]*a[2][2]
    odwrotna[1][1] = a[0][0]*a[2][2] - a[0][2]*a[2][0]
    odwrotna[1][2] = a[0][2]*a[1][0] - a[0][0]*a[1][2]

    odwrotna[2][0] = a[1][0]*a[2][1] - a[1][1]*a[2][0]
    odwrotna[2][1] = a[0][1]*a[2][0] - a[0][0]*a[2][1]
    odwrotna[2][2] = a[0][0]*a[1][1] - a[0][1]*a[1][0]

    return odwrotna / det



print("z funkcji: \n", np.linalg.inv(a))
print("\n moja: \n", odwrotna_3x3(a))

# print("\n z funkcji: \n", np.linalg.inv(d)) WYZNACZIK WYNOSI 0
# print("\n moja: \n", odwrotna_3x3(d)) WYZNACZIK WYNOSI 0
```
