##### Posiadając poniższe układy równań napisz funkcję, która obliczy wartości x, y, z
###### Jedna stworzona funkcja ma rozpatrywać układ równań z 2 i 3 niewiadomymi.
** do obliczania wyznacznika użyj funkcji zaimplementowanej przez siebie, nie wbudowanej z NumPy.


######
a)
3x + 5y = -7
x + 4y = -14


b)
x + 2y + 3z = -5
3x + y – 3z = 4
-3x + 4y + 7z = -7

```python
import copy

import numpy as np

def uklad_rownan(x1=None, y1=None, z1=None, wynik1=None, x2=None, y2=None, z2=None, wynik2=None, x3=None, y3=None, z3=None, wynik3=None, ):
    count = 0
    for arg in [x1, y1, z1, wynik1, x2, y2, z2, wynik2, x3, y3, z3, wynik3]:
        if arg is not None:
            count += 1

    if count == 6:
        macierz2x2_1 = np.array([[x1, y1],
                               [x2, y2]])
        wyznacznik2_2_1 = macierz2x2_1[0][0] * macierz2x2_1[1][1] - macierz2x2_1[0][1] * macierz2x2_1[1][0]

        macierz2x2_2 = copy.copy(macierz2x2_1)
        macierz2x2_2[0][0] = wynik1
        macierz2x2_2[1][0] = wynik2
        wyznacznik2_2_2 = macierz2x2_2[0][0] * macierz2x2_2[1][1] - macierz2x2_2[0][1] * macierz2x2_2[1][0]

        macierz2x2_3 = copy.copy(macierz2x2_1)
        macierz2x2_3[0][1] = wynik1
        macierz2x2_3[1][1] = wynik2
        wyznacznik2_2_3 = macierz2x2_3[0][0] * macierz2x2_3[1][1] - macierz2x2_3[0][1] * macierz2x2_3[1][0]

        odp_x = wyznacznik2_2_2/ wyznacznik2_2_1
        odp_y = wyznacznik2_2_3/ wyznacznik2_2_1

        return odp_x, odp_y
    elif count == 12:
        macierz_3x3_1 = np.array([[x1, y1, z1],
                                  [x2, y2, z2],
                                  [x3, y3, z3]])

        if macierz_3x3_1.shape[0] != macierz_3x3_1.shape[1]:
            raise ValueError("Podana macierz nie jest kwadratowa")
        elif macierz_3x3_1.shape[0] != 3 or macierz_3x3_1.shape[1] != 3:
            raise ValueError("Powstała macierz nie jest macierzą 3 x 3")
        else:
            wyzniacznik3_3_1 = macierz_3x3_1[0][0]*macierz_3x3_1[1][1]*macierz_3x3_1[2][2] + macierz_3x3_1[0][1]*macierz_3x3_1[1][2]*macierz_3x3_1[2][0] + macierz_3x3_1[0][2]*macierz_3x3_1[1][0]*macierz_3x3_1[2][1] - macierz_3x3_1[0][2]*macierz_3x3_1[1][1]*macierz_3x3_1[2][0] - macierz_3x3_1[0][0] * macierz_3x3_1[1][2]*macierz_3x3_1[2][1] - macierz_3x3_1[0][1]*macierz_3x3_1[1][0]*macierz_3x3_1[2][2]

        if(wyzniacznik3_3_1 == 0):
            raise ValueError("Wyznacznik macierzy jest równy 0!")

        macierz_3x3_2 = copy.copy(macierz_3x3_1)
        macierz_3x3_2[0][0] = wynik1
        macierz_3x3_2[1][0] = wynik2
        macierz_3x3_2[2][0] = wynik3
        wyzniacznik3_3_2 = macierz_3x3_2[0][0]*macierz_3x3_2[1][1]*macierz_3x3_2[2][2] + macierz_3x3_2[0][1]*macierz_3x3_2[1][2]*macierz_3x3_2[2][0] + macierz_3x3_2[0][2]*macierz_3x3_2[1][0]*macierz_3x3_2[2][1] - macierz_3x3_2[0][2]*macierz_3x3_2[1][1]*macierz_3x3_2[2][0] - macierz_3x3_2[0][0] * macierz_3x3_2[1][2]*macierz_3x3_2[2][1] - macierz_3x3_2[0][1]*macierz_3x3_2[1][0]*macierz_3x3_2[2][2]

        macierz_3x3_3 = copy.copy(macierz_3x3_1)
        macierz_3x3_3[0][1] = wynik1
        macierz_3x3_3[1][1] = wynik2
        macierz_3x3_3[2][1] = wynik3
        wyzniacznik3_3_3 = macierz_3x3_3[0][0]*macierz_3x3_3[1][1]*macierz_3x3_3[2][2] + macierz_3x3_3[0][1]*macierz_3x3_3[1][2]*macierz_3x3_3[2][0] + macierz_3x3_3[0][2]*macierz_3x3_3[1][0]*macierz_3x3_3[2][1] - macierz_3x3_3[0][2]*macierz_3x3_3[1][1]*macierz_3x3_3[2][0] - macierz_3x3_3[0][0] * macierz_3x3_3[1][2]*macierz_3x3_3[2][1] - macierz_3x3_3[0][1]*macierz_3x3_3[1][0]*macierz_3x3_3[2][2]

        macierz_3x3_4 = copy.copy(macierz_3x3_1)
        macierz_3x3_4[0][2] = wynik1
        macierz_3x3_4[1][2] = wynik2
        macierz_3x3_4[2][2] = wynik3
        wyzniacznik3_3_4 = macierz_3x3_4[0][0]*macierz_3x3_4[1][1]*macierz_3x3_4[2][2] + macierz_3x3_4[0][1]*macierz_3x3_4[1][2]*macierz_3x3_4[2][0] + macierz_3x3_4[0][2]*macierz_3x3_4[1][0]*macierz_3x3_4[2][1] - macierz_3x3_4[0][2]*macierz_3x3_4[1][1]*macierz_3x3_4[2][0] - macierz_3x3_4[0][0] * macierz_3x3_4[1][2]*macierz_3x3_4[2][1] - macierz_3x3_4[0][1]*macierz_3x3_4[1][0]*macierz_3x3_4[2][2]



        odp_x = wyzniacznik3_3_2/wyzniacznik3_3_1
        odp_y = wyzniacznik3_3_3/wyzniacznik3_3_1
        odp_z = wyzniacznik3_3_4/wyzniacznik3_3_1

        return odp_x, odp_y, odp_z

    else:
        raise ValueError("Podałeś niewłaściwą ilość współczynników!")


print(f"Dla układu równań: \n 3x + 5y = -7 \n x + 4y = -14 \n\n x wynosi: {uklad_rownan(x1=3, y1=5, wynik1=-7, x2=1, y2=4, wynik2=-14)[0]}\n y wynosi: {uklad_rownan(x1=3, y1=5, wynik1=-7, x2=1, y2=4, wynik2=-14)[1]}")
print("-------------------------------------------------------------------------------------------------------------------")
print(f"Dla układu równań: \n x + 2y + 3z = -5 \n 3x + y - 3z = 4\n-3x + 4y + 7z = -7 \n\nx wynosi: {uklad_rownan(x1=1, y1=2, z1=3, wynik1=-5, x2=3, y2=1 ,z2=-3, wynik2=4, x3=-3, y3=4, z3=7, wynik3=-7)[0]}\ny wynosi: {uklad_rownan(x1=1, y1=2, z1=3, wynik1=-5, x2=3, y2=1 ,z2=-3, wynik2=4, x3=-3, y3=4, z3=7, wynik3=-7)[1]}\nz wynosi: {uklad_rownan(x1=1, y1=2, z1=3, wynik1=-5, x2=3, y2=1 ,z2=-3, wynik2=4, x3=-3, y3=4, z3=7, wynik3=-7)[2]}" ,
      )
```
